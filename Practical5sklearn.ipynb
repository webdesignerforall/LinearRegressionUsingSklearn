```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('california_housing_train.csv')
```

```python
df
```

```python
pop=df["population"]
bedrooms=df["total_bedrooms"]
pop
```

```python
bedrooms
```

```python
pop.shape
```

```python
bedrooms.shape
```

```python
plt.scatter(pop,bedrooms)
plt.xlabel('Population')
plt.ylabel('Total Bedrooms')
plt.show()
```

```python
pop_np=pop.to_numpy() 
```

```python
bedrooms_np=bedrooms.to_numpy()
```

```python
pop_np.shape
```

```python
bedrooms_np.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
sklearn_model=LinearRegression().fit(pop_np.reshape((17000,1)),bedrooms_np)
```

```python
bedrooms_prediction=sklearn_model.predict(pop_np.reshape((17000,1)))
```

```python
plt.scatter(pop,bedrooms)
plt.xlabel('Population')
plt.ylabel('Total Bedrooms')
plt.plot(pop,bedrooms_prediction)
plt.show()
```
